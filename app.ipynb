```python
from langchain.document_loaders import TextLoader
loader = TextLoader("data.txt")
document=loader.load()
print(document)
```

```python
import textwrap

def warp_text_preserve_newlines(text, width=110):
    lines = text.split('\n')
    wraped_lines = [textwrap.fill(line,width=width) for line in lines]
    wrapped_text = '\n'.join(wraped_lines)

    return wrapped_text

```

```python
print(warp_text_preserve_newlines(str(document[0])))
```

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(document)
print(docs[0])
print(len(docs))

```

```python
from langchain.embeddings import HuggingFaceEmbeddings
import os
from langchain.vectorstores import FAISS
os.environ['HUGGINGFACEHUB_API_TOKEN'] = "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"
embeddings = HuggingFaceEmbeddings()
doc = FAISS.from_documents(docs, embeddings)

```

```python
query = "What is AIUB?"
db = doc.similarity_search(query)
print(warp_text_preserve_newlines(str(db[0].page_content)))
```

```python
from langchain.chains.question_answering import load_qa_chain
from langchain import HuggingFaceHub
llm = HuggingFaceHub(repo_id="google/flan-t5-base", model_kwargs={"temperature":0.8, "max_length":512})

chain = load_qa_chain(llm, chain_type="stuff")
query2 = "What is the full meaning of AIUB?"
doc2= doc.similarity_search(query2)
chain.run(input_documents = doc2, question = query2)
```
